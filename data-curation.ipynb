```python
import json

# Path to your JSON lines file (or wherever you have the data)
json_path = '/kaggle/input/listing/listings/metadata/listings_3.json'

# To store filtered entries
filtered_entries = []
linenum = 0
# Read and parse
with open(json_path, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        try:
            data = json.loads(line)
            linenum += 1
            # Check if 'brand' exists and has at least one item
            if 'brand' in data and isinstance(data['brand'], list):
                for brand_entry in data['brand']:
                    if 'language_tag' in brand_entry and brand_entry['language_tag'].startswith('en'):
                        filtered_entries.append(data)
                        break  # Once matched, no need to check other brand entries for this item

        except json.JSONDecodeError as e:
            print(f"Skipping invalid line: {e}")

print(f"Total entries with brand language_tag starting with 'en' : {len(filtered_entries)} and total are {linenum}")
```

```python
import os
import json
from tqdm import tqdm  # for progress bar

# Root folder where your extracted dataset is
root_folder = '/kaggle/input/listing/listings/metadata'
unique_language = set()
# To store matching entries
filtered_entries = []

# Walk through all folders and files
for dirpath, dirnames, filenames in os.walk(root_folder):
    for filename in filenames:
        if filename.endswith('.json'):  # Only process JSON files
            json_path = os.path.join(dirpath, filename)
            filterhere = []
            linenum = 0
            # Open each JSON file
            with open(json_path, 'r', encoding='utf-8') as f:
                for line in tqdm(f, desc=f"Processing {filename}"):
                    line = line.strip()
                    if not line:
                        continue
                    try:
                        linenum += 1
                        data = json.loads(line)

                        # Check for brand and English language_tag
                        if 'brand' in data and isinstance(data['brand'], list):
                            for brand_entry in data['brand']:
                                if 'language_tag' in brand_entry and brand_entry['language_tag'].startswith('en'):
                                    filtered_entries.append(data)
                                    filterhere.append(data)
                                    unique_language.add(brand_entry['language_tag'])
                                    break  # no need to check other brands for this item

                    except json.JSONDecodeError as e:
                        print(f"Skipping invalid JSON line in {filename}: {e}")
            print(f"Total entries with brand language_tag starting with 'en' : {len(filterhere)} and total are {linenum}")

print(f"\n✅ Total entries collected: {len(filtered_entries)}")


# Save filtered data to a new JSONL file
save_path = '/kaggle/working/filtered_entries.jsonl'
with open(save_path, 'w', encoding='utf-8') as f:
    for entry in filtered_entries:
        f.write(json.dumps(entry) + '\n')  # Write each JSON object on a new line

print(f"✅ Filtered entries saved at: {save_path}")
```

```python
import json

input_path = '/kaggle/working/filtered_entries.jsonl'  # change your file path here
output_path = '/kaggle/working/filtered_entries_2.jsonl'

filtered_entries = []

# Open the file
with open(input_path, 'r', encoding='utf-8') as f:
    # If your file has multiple lines, each a JSON object
    for line in f:
        line = line.strip()
        if not line:
            continue
        try:
            data = json.loads(line)

            # List of fields that can have 'language_tag'
            keys_to_check = ['brand', 'bullet_point', 'item_name', 'color', 'material', 'item_keywords', 'model_name', 'style']

            for key in keys_to_check:
                if key in data:
                    filtered_list = [
                        entry for entry in data[key]
                        if 'language_tag' in entry and entry['language_tag'].startswith('en')
                    ]
                        
                    if filtered_list:
                        data[key] = filtered_list
                        for entry in data[key]:
                            if 'language_tag' in entry:
                                del entry['language_tag']   
                        
                        
                    else:
                        del data[key]
            id_fields = ['other_image_id','item_id', 'model_number', 'marketplace','domain_name','3dmodel_id','spin_id']
            for id_field in id_fields:
                if id_field in data:
                    del data[id_field]
            if 'node' in data:
                for entry in data['node']:
                    del entry['node_id']

            filtered_entries.append(data)

        except json.JSONDecodeError as e:
            print(f"Skipping invalid JSON line: {e}")

print(f"\n✅ Total entries collected: {len(filtered_entries)}")


```
